## 1. Introdução à Machine Learning

Depois das etapas de preparação e análise dos dados, aplicamos algoritmos de aprendizado de máquina para criar modelos que possam identificar padrões e fazer previsões. Nesta seção, focaremos no **Supervised Learning**.

### O que é Supervised Learning?  
Supervised Learning (aprendizado supervisionado) é uma abordagem onde o modelo é treinado em um dataset com **variáveis de entrada (features)** e uma **variável alvo (target)** conhecida. O objetivo é que o modelo aprenda a mapear as features para prever o target de forma precisa em dados novos.

### Em outras palavras
Aprendizado supervisionado é aquele que se conhece a resposta esperada, tendo como objetivo aprender a mapear os atributos para essa resposta. Essa resposta esperada é comumente chamada de target.




### Exemplo
Exemplo da tabela do Titanic: tenho um conjunto de atributos, e, a partir destes atributos, posso prever o target. A partir da idade, sexo etc, tentamos prever se o tripulante vai morrer ou sobreviver. 



------

## Existem dois tipos de aprendizado supervisionado:

- Classificação (duas ou mais classes)
Target é um valor discreto ou categórico (ex: doente ou saudável, permitido ou negado, labrador, salsicha ou doberman pinscher).

- Regressão (queremos prever um valor contínuo, um número)
Target é um valor real ou número (ex: preço de uma casa, volume de água).
