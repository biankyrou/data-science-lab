# Modelos de Machine Learning
Modelos de Machine Learning (ML) são algoritmos que permitem aos computadores aprenderem padrões e realizarem previsões a partir de dados. Esses modelos são divididos em duas categorias principais: supervisionados, onde há dados rotulados (ex.: classificação e regressão), e não supervisionados, que identificam padrões ocultos em dados não rotulados (ex.: clustering).


1. K-Nearest Neighbors (KNN)
KNN é um algoritmo de aprendizado supervisionado usado para classificação e regressão. Ele funciona encontrando os "K" vizinhos mais próximos de um dado com base em uma medida de distância (como Euclidiana) e usando a maioria para classificar ou calcular a média para prever valores. KNN é simples, mas pode ser lento com grandes datasets porque compara cada novo ponto com todos os pontos de treinamento.

    

2. Regressão Linear
A regressão linear é um método para modelar a relação entre uma variável dependente (y) e uma ou mais variáveis independentes (X). Ela busca a melhor linha reta que minimiza a diferença entre os valores previstos e os reais. É muito utilizada para prever valores contínuos e entender como as variáveis explicativas afetam a variável alvo.

    

3. Árvore de Decisão
A árvore de decisão é um algoritmo que divide dados em subconjuntos com base em perguntas sequenciais sobre as características. Cada decisão gera um nó que representa um possível resultado, até chegar a uma folha (resposta final). Ela é fácil de interpretar e usada tanto para classificação quanto para regressão, mas pode se tornar muito complexa e se ajustar demais aos dados (overfitting).
