## Avaliação
A princípio um problema de Machine Learning possui dois conjuntos de dados:
- O conjunto de treino (onde o target é conhecido). Vou usar para treinar, portanto, preciso das respostas.
- O conjunto de teste (onde não conhecemos o target e queremos descobri-lo).

### Como vamos avaliar o funcionamento do modelo?
- É comum dividir o conjunto de treino em um conjunto de validação (validation set) e o conjunto de treino que o modelo irá receber.
- Assim, temos as respostas corretas e podemos compará-las com as produzidas pelo nosso modelo.

------

## Validation Set
Exemplo: Se tenho 1000 perguntas e respostas que já sei, eu vou treinar o algoritmo com 800 apenas. As outras 200 vão ser o conjunto de validação. Portanto, o conjunto de validação são respostas que eu já sei, mas que meu modelo nunca viu.




### Primeiro momento:
- Training set: vou ter um conjunto de treino, que eu sei a resposta (atributos e targets).
- Test set: só tenho os atributos, sem as respostas.

→ Importante que o algoritmo entenda e aprenda os padrões, não que ele os decore.

### Segundo momento:
- Vamos ter o val set: um conjunto de atributos que o algoritmo não sabe a resposta. Ele vai ter acesso somente ao training set.

------

## Em outras palavras

### Divisão de Dados em Treino, Validação e Teste

#### Treino (Training Set)
- É a parte dos dados que você usa para ensinar o modelo.
- O modelo aprende a reconhecer padrões a partir desses dados.

#### Validação (Validation Set)
- É um subconjunto dos dados que você usa para ajustar o modelo.
- Você avalia o desempenho do modelo durante o treinamento e faz ajustes, como escolher hiperparâmetros, para melhorar a precisão.

#### Teste (Test Set)
- É a parte dos dados que você usa para avaliar a performance final do modelo.
- Esses dados não são usados durante o treinamento e ajudam a verificar se o modelo generaliza bem para novos dados.

#### Divisão Simples
- **Total de Dados**: Imagine que você tem 100 dados.
- **Divisão**: Você pode usar 70% (70 dados) para treino, 15% (15 dados) para validação e 15% (15 dados) para teste.
